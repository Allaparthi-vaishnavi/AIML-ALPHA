```python

import statsmodels.formula.api as smf
model1 = smf.ols("sunday~daily", data = data1).fit()
```

```python
model1.summary()
```

# The predicted equation is 
*y^ = Beta_0+Beta_1x

*y^ = 13.8356 + 1.3397x

*The probability(p-value) for intercept (beta_0) is 0.707 > 0.05.

*Therfore the intercept coefficient may not be that much significant in prediction.

*However the p-value for "daily" (beta_1) is 0.00 < 0.05.

*Therefore the beta_1 coeeficient is highly significant and is contributint to prediction.

```python
# Predict on all given training data

pred = model1.predict(data1["daily"])
pred
```

```python
# Add Predicted values as a column in data1

data1["Y_hat"]=pred
data1
```

```python
# Compute the error values (residuals) and add as another column

data1["residuals"]=data1["sunday"]-data1["Y_hat"]
data1
```

```python
# Compute Mean Squared Error for the model

mse = np.mean((data1["daily"]-data1["Y_hat"])**2)
rmse = np.sqrt(mse)
print("MSE: ", mse)
print("RMSE: ", rmse)
```

```python
# Plot the residuals versus y_hat (to check wheather residuals are independent of each)

plt.scatter(data1["Y_hat"], data1["residuals"])
```

Observations:

*The residuals data point appears to be no trend and the residuals are randomly placed around the zero error line.

*Hence the assupmtion of homoscedasticty is stisfied (constant varinace in residuals)

```python
# Plot the Q-Q plot(to check the normality of residuals)

import statsmodels.api as sm
sm.qqplot(data1["residuals"], line='45', fit=True)
plt.show()
```

```python
sns.histplot(data1["residuals"], kde=True)
```

Observations:

*The data points are seen to closely follow the reference line of normality.

*Hence the residuals are approximately normally distributed as also can be seen from the histplot from the kde graph.
