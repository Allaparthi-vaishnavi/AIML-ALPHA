```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset
df = pd.read_csv('Pharma_data.csv')  # Replace 'data.csv' with the actual dataset filename

# Data Cleaning
print("Missing Values:\n", df.isnull().sum())
print("Duplicates:", df.duplicated().sum())
df.drop_duplicates(inplace=True)
df.dropna(inplace=True)

# Bar Plot: Total Sales per Region
plt.figure(figsize=(10, 6))
sns.barplot(x='Region', y='Sales', data=df, estimator=sum, ci=None)
plt.title('Total Sales per Region')
plt.xticks(rotation=45)
plt.show()

# Scatter Plot: Marketing Spend vs Sales
plt.figure(figsize=(8, 5))
sns.scatterplot(x='Marketing_Spend', y='Sales', data=df)
plt.title('Marketing Spend vs Sales')
plt.xlabel('Marketing Spend')
plt.ylabel('Sales')
plt.show()

# Boxplot: Drug Effectiveness Across Age Groups
plt.figure(figsize=(10, 6))
sns.boxplot(x='Age_Group', y='Effectiveness', data=df)
plt.title('Drug Effectiveness Across Age Groups')
plt.show()

# Line Plot: Sales Trend Over Trial Periods
plt.figure(figsize=(12, 6))
sns.lineplot(x='Trial_Period', y='Sales', hue='Product', data=df)
plt.title('Sales Trend Over Trial Periods')
plt.show()

# Heatmap: Correlation Between Sales, Marketing Spend, and Effectiveness
plt.figure(figsize=(8, 5))
corr_matrix = df[['Sales', 'Marketing_Spend', 'Effectiveness']].corr()
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Heatmap')
plt.show()

# Observations and Insights
print("Observations:")
print("1. Regional sales distribution varies, with some regions outperforming others.")
print("2. A positive correlation is expected between marketing spend and sales.")
print("3. Some age groups may show higher drug effectiveness based on boxplot insights.")
print("4. The line plot will highlight sales trends across trial periods for different products.")
print("5. Correlation heatmap will show interdependence between sales, marketing spend, and effectiveness.")
```
