```python
import pandas as pd
import matplotlib.pyplot as plt

# Sample DataFrame
data = {
    "Account_Type": ["Savings", "Current", "Savings", "Current", "Business", "Savings", "Business", "Current", "Savings", "Business"],
    "Transaction_Amount": [200, 500, 300, 400, 250, 600, 150, 100, 450, 50],
    "Branch": ["A", "A", "B", "B", "A", "C", "C", "B", "C", "A"]
}

df = pd.DataFrame(data)
df
```

```python
# Group data by Account_Type and sum Transaction_Amount
account_type_sum = df.groupby("Account_Type")["Transaction_Amount"].sum()

# Plot the bar chart
plt.figure(figsize=(8, 5))
account_type_sum.plot(kind="bar", color="skyblue", edgecolor="black")
plt.title("Total Transaction Amount per Account Type")
plt.xlabel("Account Type")
plt.ylabel("Total Transaction Amount")
plt.xticks(rotation=45)
plt.grid(axis="y", linestyle="--", alpha=0.7)
plt.show()
```

```python
# Count transactions per Branch
branch_counts = df["Branch"].value_counts()

# Plot the pie chart
plt.figure(figsize=(6, 6))
branch_counts.plot(kind="pie", autopct="%1.1f%%", colors=["lightcoral", "lightblue", "lightgreen"])
plt.title("Percentage of Transactions per Branch")
plt.ylabel("")  # Hide y-label
plt.show()
```
